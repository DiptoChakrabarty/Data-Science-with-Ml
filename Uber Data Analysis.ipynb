```
import pandas as pd
import seaborn
import pylab
```

```
df=pd.read_csv("https://github.com/fivethirtyeight/uber-tlc-foil-response/blob/master/uber-trip-data/uber-raw-data-apr14.csv?raw=true")
df.head(5)
```

```
dt=df['Date/Time']
d,t=dt[0].split()
print(d)
print(t)
```

```
dy,m,yr=d.split('/')
print(dy)
print(m)
print(yr)
```

```
df['Date/Time']=pd.to_datetime(df['Date/Time'])
```

```
def get_day(dt):
  return dt.day
df['day']=df['Date/Time'].map(get_day)
```

```
df.head(5)
```

```
def weekday(dt):
  return dt.weekday()
df['weekday']=df['Date/Time'].map(weekday)
```

```
def hour(dt):
  return dt.hour
df['Hour']=df['Date/Time'].map(hour)
```

```
df.head(5)
```

```
import matplotlib.pyplot  as plt
%matplotlib inline
```

```
plt.hist(df.day, bins=30, rwidth=.8, range=(0.5, 30.5))
plt.xlabel('Day')
plt.ylabel('Frequency')
plt.title('Frequency by Day - uber - Apr 2014')
display()
```

```
k=df.groupby('day')
for i,j in k:
  print(i,len(j))
```

```
k=df.groupby('day')
for i,j in k:
  print(i,j)
```

```
def count_rows(i):
  return len(i)

by_day=df.groupby('day').apply(count_rows)
by_day
```

```
plt.bar(range(1,31),by_day)
display()
```

```
day_sort=by_day.sort_values()
day_sort
```

```
plt.bar(range(1,31),day_sort,color='m')
plt.xticks(range(1,31), day_sort.index)
plt.xlabel('date of the month')
plt.ylabel('frequency')
plt.title('Frequency by Day - uber - Apr 2014')
display()
```
