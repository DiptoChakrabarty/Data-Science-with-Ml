```
import pandas as pd
df = pd.read_csv('https://github.com/QuantEcon/QuantEcon.lectures.code/raw/master/pandas/data/ticker_data.csv')
```

```
da=df.set_index('Date')
da.head(5)
```

```
da.describe()
```

```
ticker_list = {'INTC': 'Intel',
               'MSFT': 'Microsoft',
               'IBM': 'IBM',
               'BHP': 'BHP',
               'TM': 'Toyota',
               'AAPL': 'Apple',
               'AMZN': 'Amazon',
               'BA': 'Boeing',
               'QCOM': 'Qualcomm',
               'KO': 'Coca-Cola',
               'GOOG': 'Google',
               'SNE': 'Sony',
               'PTR': 'PetroChina'}
```

```
l=pd.Series()
```

```
for i in ticker_list:
  p= ((da.loc[da.index[-1],i]-da.loc[da.index[0],i])/da.loc[da.index[0],i])*100
  n=ticker_list[i]
  l[n]=p
```

```
l
```

```
l.sort_values(inplace=True)
```

```
import matplotlib.pyplot as plt
```

```
fig, ax = plt.subplots(figsize=(10,8))
l.plot(kind='bar', ax=ax)
plt.show()
display()
```
