```
import pandas as pd
realwage = pd.read_csv('https://github.com/QuantEcon/QuantEcon.lectures.code/raw/master/pandas_panel/realwage.csv')
```

```
new_table=realwage.pivot_table(index='Time',values='value',columns=['Country','Pay period','Series'])
type(new_table.index)
```

```
new_table.tail(5)
```

```
new_table.columns.names
```

```
new_table.index=pd.to_datetime(new_table.index)
type(new_table.index)
type(new_table.columns)
```

```
new_table.stack().head()
```

```
new_table.head()
```

```
new_table.stack(level='Country').head()
```

```
new_table['2016']
```

```
final = new_table.xs(('Hourly', 'In 2015 constant prices at 2015 USD exchange rates'),
                         level=('Pay period', 'Series'), axis=1)
final.head()
```

```
final.transpose().head()
#reverses the columns and row positions
```

```
worlddata = pd.read_csv('https://github.com/QuantEcon/QuantEcon.lectures.code/raw/master/pandas_panel/countries.csv', sep=';')
worlddata.columns
```

```
worlddata = worlddata[['Country (en)', 'Continent']]
worlddata = worlddata.rename(columns={'Country (en)': 'Country'})
worlddata.head()
#Rename the column names
```

```
merge=pd.merge(final.transpose(),worlddata,how="left",left_index=True,right_on='Country')
merge.head()
```

```
merge[merge['Continent'].isnull()]
#countries present in final data but not in world data have values NULL
```

```
missing_continents = {'Korea': 'Asia',
                      'Russian Federation': 'Europe',
                      'Slovak Republic': 'Europe'}
merge['Continent']=merge['Continent'].fillna(merge['Country'].map(missing_continents))
#replace the missing continents by values of your own
```

```
merge[merge['Country']=='Korea']
#check for Korea 
```

```
r=['Central America', 'North America', 'South America']
for i in r:
  merge['Continent'].replace(to_replace=i,value='America',inplace=True)
  
merge.head(5)
```

```
merge.transpose().head()
```

```
merge.head()
```

```
merged=merge.set_index(['Continent','Country']).sort_index()
merged.head()
```

```
merged.columns = pd.to_datetime(merged.columns)
merged.columns = merged.columns.rename('Time')
merged.columns
```

```
merged.index
```

```
merged=merged.transpose()
```

```
merged.mean()
```

```
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```
merged.mean().sort_values(ascending=False).plot(kind='bar', title="Average real minimum wage 2006 - 2016")

#Set country labels
country_labels = merged.mean().sort_values(ascending=False).index.get_level_values('Country').tolist()
plt.xticks(range(0, len(country_labels)), country_labels)
plt.xlabel('Country')

plt.show()
```
