```
mycol=df.select('Survived',
 'Pclass',
 'Name',
 'Sex',
 'Age',
 'SibSp',
 'Parch',
 'Ticket',
 'Fare','Embarked')
```

```
final=mycol.na.drop()
#removes missing data
```

```
from pyspark.ml.feature import VectorAssembler,VectorIndexer,OneHotEncoder,StringIndexer
```

```
indexer=StringIndexer(inputCol='Sex',outputCol='index')
gen=OneHotEncoder(inputCol="index",outputCol='sexvec')
```

```
emindexer=StringIndexer(inputCol='Embarked',outputCol='emindex')
emcoder=OneHotEncoder(inputCol="emindex",outputCol='emvec')
```

```
assembler=VectorAssembler(inputCols=['Pclass','sexvec','emvec','Age','SibSp','Parch','Fare'],outputCol='features')
```

```
from pyspark.ml import Pipeline
```

```
logreg=LogisticRegression(featuresCol='features',labelCol='Survived')
```

```
pipeline = Pipeline(stages=[indexer,emindexer,
                               gen,emcoder,
                               assembler,logreg])
```

```
train,test=final.randomSplit([0.65,0.35])
```

```
fitmod=pipeline.fit(train)
```

```
results=fitmod.transform(test)
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```
eval=BinaryClassificationEvaluator(rawPredictionCol='prediction',labelCol='Survived')
```
