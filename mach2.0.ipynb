```
from pyspark.sql import SparkSession
```

```
spark=SparkSession.builder.appName('mach2.0').getOrCreate()
from pyspark.ml.regression import LinearRegression
wa=sqlContext.sql("SELECT * FROM ecom")
```

```
wa.show()
```

```
wa.printSchema()
```

```
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
```

```
wa.columns
```

```
assembler=VectorAssembler(inputCols=['Avg Session Length','Time on App','Time on Website','Length of Membership'],outputCol='features')
```

```
output = assembler.transform(wa)
```

```
output.printSchema()
```

```
output.select('features').show()
```

```
final=output.select('features','Yearly Amount Spent')
```

```
final.show()
```

```
train_data,test_data=final.randomSplit([0.7,0.3])
```

```
train_data.describe().show()
```

```
p,q=final.randomSplit([0.6,0.4])
```

```
lr=LinearRegression(labelCol='Yearly Amount Spent',featuresCol='features')
```

```
model=lr.fit(train_data)
```

```
tres=model.evaluate(test_data)
```
